```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Titanic.csv')
data.head()
```

```python
data.dtypes
```

```python
data = data.drop(["PassengerId", "Name", "Ticket", "SibSp", "Embarked", "Parch"], axis = 1)
data.head()
```

```python
data["Age"] = 1912 - data["Age"]
data.head()
```

```python
data = data.rename(columns = \
                    {'Survived' : 'Выжил', \
                    'Pclass' : 'Класс',\
                    'Sex' : 'Пол',\
                    'Age' : 'Год_рождения',\
                    'Fare' : 'Цена',\
                    'Cabin' : 'Каюта'})
data.head()
```

```python
data.loc[:,'Каюта'] = data.loc[:,'Каюта'].fillna('C85')
data
```
